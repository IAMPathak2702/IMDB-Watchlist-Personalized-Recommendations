```python
book_rating_df = ratings.merge(books,on="ISBN")
```

```python
num_rating_df = book_rating_df.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating_df.rename(columns={'Book-Rating' : "num_rating"},inplace=True)
num_rating_df
```

```python
avg_rating_df = book_rating_df.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': "avg_rating"}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values("avg_rating", ascending=False)
```

```python
popular_df
```

```python
dff =popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "avg_rating",'Book-Author','Year-Of-Publication','Image-URL-S']]
```

```python
dff.to_csv("../src/data/popular_df.csv",index=False )
```
